```python
# 「array」のリストを作成
array = ["月曜日は晴れです", "火曜日は雨です", "水曜日は晴れです", "木曜日は晴れです", "金曜日は曇りです", "土曜日は曇りのち雨です", "日曜日は雷雨です"]

# 確認のための出力
print(array)
```

```python
# 「dictionary」という名前のディクショナリを作成
dictionary = {"mon": "晴れ", "tue": "雨", "wed": "晴れ", "thu": "晴れ", "fri": "曇り", "sat": "曇りのち雨", "sun": "雷雨"}

# 確認のための出力
print(dictionary)
```

```python
# リストから水曜日の天気予報を取り出して出力
print(array[2])

# ディクショナリから水曜日の天気予報を取り出して出力
print(dictionary["wed"])
```
