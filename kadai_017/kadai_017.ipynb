```python
class Human:
  def __init__(self, name, age):
      self.name = name
      self.age = age

  def check_adult(self):
      if self.age >= 20:
          print(f"{self.name}さんは大人です。（年齢：{self.age}）")
      else:
          print(f"{self.name}さんは大人ではありません。（年齢：{self.age}）")

human_list = [Human("侍太郎", 35), Human("侍花子", 16), Human("侍二郎", 23)]

for human in human_list:
      human.check_adult()
```
